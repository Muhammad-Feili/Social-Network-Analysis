این میتواند به عنوان بررسی اینکه کدام گرو ها دارای بیشترین مباحث داغ هستند یا اینکه بررسی کاربرانی که بیشترین فعالیت

# توضیحات مربوط به تمرین سوم در خصوص تعریف یک مسءله باز 

چنین تعریف کرد(subreddits) می توان از این گراف به منظور رابطه ی بین گروه ها
ابتدا تعداد کاربرانی که به طور مشترک در دو گروه فعالیت دارند را به دست آورد  و همچنین نحوه تعریف کاربران فعال در
دو گروه به این صورت است که اگر کاربر بیش از 10 پست در همان گروه گذاشته باشد آن کاربر فعال نامیده می شود.
در این مسءله وزن یال ها همان تعداد کاربرانی است که در دو گروه یا همان راس های یال مورد نظر فعال هستند!





```python
import pandas as pd
data=pd.read_csv('data.csv')
data.head()
```

```python
data=data.as_matrix()
```

```python
DATA=dict()
for edge in data:
    for EDGE in data:
        if edge[0]!=EDGE[0] and edge[1]==EDGE[1]:
            if (edge[0],EDGE[0]) not in DATA.keys() or (EDGE[0], edge[0]) not in DATA.keys():
                DATA[(edge[0],EDGE[0])]=1
            else:
                DATA[(edge[0],EDGE[0])]+=1
```

```python
len(DATA)
```

```python
import networkx as nx

G=nx.Graph()
    
# adding the edges of graph
for edge in DATA.keys():
    G.add_edge(edge[0], edge[1], weight=DATA[edge])
```

```python
print(nx.info(G))
```

```python
nx.spring_layout(G)
nx.draw_random(G)
```

```python
import matplotlib.pyplot as plt
 
# larger figure size
plt.figure(3,figsize=(20,20)) 
nx.draw_random(G,node_size=10, node_color='#0000ff',
    edge_color='#999999', with_labels=False)
plt.show()
```
